### Calculate the probability that $4 \leq x \leq 5$ using a scipy.stats function and compare it against a Monte Carlo estimate of the integral

\begin{equation}
\begin{aligned}
P_x(4 \leq x \leq 5) &= \int_{4}^{5} p(x) dx \\
&= \int_{-\infty}^{5} p(x)dx - \int_{-\infty}^{4} p(x) dx \\
& \text{where the CDF of } p(x) \text{ is defined as: } P(x \leq X) = F_x(X) = \int_{-\infty}^{X}p(x)dx \\
&= F_x(5) - F_x(4) \\
\end{aligned}
\end{equation}

Monte Carlo integration:
\begin{equation}
\begin{aligned}
\int_{-\infty}^{\infty}p(x)f(x)dx = \lim_{n \rightarrow \infty} \frac{1}{N} \sum_{i=1}^{N} f(x_i),
\end{aligned}
\end{equation}
where $x_i \sim p(x)$.

Let $f_4(x) = \begin{cases} 
1 & \text{if } x \leq 5, \\
0 & \text{otherwise.}
\end{cases}$ and let 

Let $f_5(x) = \begin{cases} 
1 & \text{if } x \leq 5, \\
0 & \text{otherwise.}
\end{cases}$.

Why does this work? Well, we are approximating two integrals that are 'active' up to their upper bounds. This is equivalent to setting $p(x)$ to zero at any point above the bounds. To expand on this idea, let us write this out in full:



```python
#Using the CDF
p_range_scistats = pdf_x.cdf(5) - pdf_x.cdf(4)

#Using Monte Carlo
N_monte_carlo = 10000

samples = pdf_x.rvs(N_monte_carlo)

f_4 = lambda x: 1 * (x <= 4)
f_5 = lambda x: 1 * (x <= 5)

upper_MCI = 1 / N_monte_carlo * np.sum(f_5(samples))
lower_MCI = 1 / N_monte_carlo * np.sum(f_4(samples))

p_range_MCI = upper_MCI - lower_MCI

#Visualise the MCI:
fig, ax = plt.subplots(1, 3, figsize = (12, 6))
ax = ax.flatten()
ax[0].hist(samples, bins = 50, density = True, label = "Empirical distribution")
ax[0].plot(x_range, pdf_x.pdf(x_range), label = "Analytical distribution")
ax[0].plot(x_range, f_4(x_range), label = r"$f_4(x) = I(x; -\infty, 4)$")
ax[0].plot(x_range, pdf_x.pdf(x_range) * f_4(x_range), label = r"$p(x) \times f_4(x)$")

ax[1].hist(samples, bins = 50, density = True, label = "Empirical distribution")
ax[1].plot(x_range, pdf_x.pdf(x_range), label = "Analytical distribution")
ax[1].plot(x_range, f_5(x_range), label = r"$f_5(x) = I(x; -\infty, 5)$")
ax[1].plot(x_range, pdf_x.pdf(x_range) * f_5(x_range), label = r"$p(x) \times f_5(x)$")

ax[2].plot(x_range, pdf_x.pdf(x_range) * f_4(x_range), "b-", label = r"$p(x) \times f_4(x)$")
ax[2].plot(x_range, pdf_x.pdf(x_range) * f_5(x_range), "r--", label = r"$p(x) \times f_5(x)$")
ax[2].set_title("Result of multiplying $p(x)$ with $f_4(x)$ and $f_5(x)$")
for axs in ax:
    axs.legend()
    axs.set_xlabel("x")
    axs.grid()
    
fig.tight_layout()
plt.savefig("./images/Q1a_fig5.png")
plt.show()

#Show how the MCI converges

N_repeats = 1000
N_mci = 10**(np.array([1, 2, 3, 4, 5, 6]))

MCI_expected = []
MCI_std = []

for N_monte in N_mci:
    
    MCI_estimate = [] #Only consider the final term, we could do both but it is not worth it
    
    for i in range(N_repeats):
        
        samples = pdf_x.rvs(N_monte)
        
        upper_MCI = 1 / N_monte * np.sum(f_5(samples))
        lower_MCI = 1 / N_monte * np.sum(f_4(samples))

        MCI_estimate.append(upper_MCI - lower_MCI)
    
    MCI_expected.append(np.mean(MCI_estimate))
    MCI_std.append(np.std(MCI_estimate))

MCI_expected = np.array(MCI_expected)
MCI_std = np.array(MCI_std)
MCI_lower = MCI_expected - 1 * MCI_std
MCI_upper = MCI_expected + 1 * MCI_std

plt.figure()
plt.semilogx(N_mci, MCI_expected, "k--", label = "Monte carlo expected value")
plt.fill_between(N_mci, MCI_lower, MCI_upper, color = "r", alpha = 0.6, label = r"$E_{x}\left[f_5(x)\right] - E_{x}\left[f_4(x)\right]$ $\pm$ 1 standard deviation")
plt.scatter([N_mci[-1]], [p_range_scistats], marker = "x", color = "b", label = "scipy.stats estimate")
plt.xlabel("Number of Monte carlo samples")
plt.grid()
plt.legend()
plt.savefig("./images/Q1a_fig6.png")
plt.show()

plt.figure()
plt.loglog(N_mci, np.abs(MCI_expected - p_range_scistats), label = "Error")
plt.xlabel("Number of Monte carlo samples")
plt.ylabel("Estimation error")
plt.grid()
plt.legend()
plt.savefig("./images/Q1a_fig7.png")
plt.show()

print(p_range_scistats, MCI_expected[-1])
```

### Compare the bias of the following estimators of the variance as a function of N:

\begin{equation}
\sigma_a^2 = \frac{1}{N}\sum_{n=1}^{N} \left(x_n - \tilde{\mu} \right)^2
\end{equation}

\begin{equation}
\sigma_b^2 = \frac{1}{N-1}\sum_{n=1}^{N} \left(x_n - \tilde{\mu} \right)^2
\end{equation}

Note, there are two sources of variation here. The first is in $\sigma_a^2$ and $\sigma_b^2$, and the other is in calculating the expectation of these terms. 

So I will set the expectation of estimator variation to a very high number (1000), and then investigate how increasing the number of times we repeat the experiment to calculate $\sigma_a^2$ and $\sigma_b^2$. If the estimator is unbiased, we would expect the expected value of the bias to tend to 0 and the variance to reduce as we repeat the calculation of $\sigma_a^2$ and $\sigma_b^2$ more and more.

```python
estimator_1 = lambda x, mu_ML: 1 / (len(x)) * np.sum((x - mu_ML)**2)
estimator_2 = lambda x, mu_ML: 1 / (len(x) - 1) * np.sum((x - mu_ML)**2)

def bias_calculator(n_iterations, estimator):
    #Estimate the expectation
    expect = np.zeros(int(n_iterations))
    
    for i in range(n_iterations):
        x_sample = pdf_x.rvs(n_samples)
        
        mu_ML = np.mean(x_sample)
        
        expect[i] = estimator(x_sample, mu_ML)
    
    bias = sigma**2 - np.mean(expect)
    
    return bias

for est_number, estimator in enumerate([estimator_1, estimator_2]):

    N_repeats_estimator = 1000
    N_iterations = 10**np.arange(1, 5, 1)

    bias_iter = np.zeros(len(N_iterations))
    std_bias_iter = np.zeros(len(N_iterations))

    for cnt, n_iter in enumerate(N_iterations):

        bias_i = []

        for i in range(N_repeats_estimator):
            bias = bias_calculator(n_iter, estimator)

            bias_i.append(bias)

        bias_iter[cnt] = np.mean(bias_i)
        std_bias_iter[cnt] = np.std(bias_i)

    plt.figure()
    plt.semilogx(N_iterations, bias_iter, "b--", label = "Bias - Estimator {}".format(est_number + 1))
    plt.fill_between(N_iterations, bias_iter - 3 * std_bias_iter,  bias_iter + 3 * std_bias_iter, color = "m", alpha = 0.5, label = "variance of bias")
    plt.legend()
    plt.xlabel("Number of expectation iterations")
    plt.savefig("./images/Q1c_fig{}.png".format(2 + est_number))
    plt.show()
```

### Formulate the maximum likelihood estimation problem for a second order polynomial.

The problem is as follows: Given a dataset $\mathcal{D} = \{ \mathbf{X}, \mathbf{t}\}$, where $\mathbf{X} = [x_1, \cdots, x_N]^T$ and $\mathbf{t} = [t_1, \cdots, t_N]^T$, $\mathbf{X} \in \mathbb{R}^{N \times 1}$, $\mathbf{t} \in \mathbb{R}^{N \times 1}$. We assume that the following model:

\begin{equation}
t_n = f(x_n) + \epsilon_n,
\end{equation}

where $\epsilon_n \sim \mathcal{N}(0, \sigma^2)$ and

\begin{equation}
f(x) = \omega_0 + \omega_1 \cdot x + \omega_2 \cdot x^2 + \cdots +  + \omega_K \cdot x^K,
\end{equation}

is a $K^{th}$ order polynomial function. We can reformulate $f(x)$ as
\begin{equation}
f(x) = \boldsymbol\omega^T \boldsymbol\phi(x),
\end{equation}
where $\boldsymbol\omega \in \mathbb{R}^{K + 1}$ and $\boldsymbol\phi(x)$ is a basis vector in the form
\begin{equation}
\boldsymbol\phi(x_n) = [\phi_{n}^{(0)}, \phi_{n}^{(1)}, \cdots, \phi_{n}^{(K)}]= [x^{0}, x^{1}, \cdots, x^{K}]^T,
\end{equation}
where $\boldsymbol\phi(\cdot) \in \mathbb{R}^{K + 1}$ and $\phi_{n}^{(i)}$ refers to the $i^{th}$ index for the $n^{th}$ input $\phi_{n}^{(i)} = \boldsymbol\phi(x_n)[i]$.

As the noise is Gaussian, the addition of $f(x)$ to $\epsilon$ gives a mean-shifted Gaussian distribution:
\begin{equation}
p(t\vert x) = \mathcal{N}(t \vert f(x), \sigma^2).
\end{equation}

In order to derive an estimator for this model, we assume that the samples in $\mathcal{D}$ satisfy the _i.i.d_ assumption. As such, we can calculate the conditional likelihood function $p(T \vert X)$ as follows
\begin{equation}
p(T \vert X) = \prod_{n=1}^{N} p(t_n \vert x_n),
\end{equation}

However, to avoid numerical underflow issues, and to improve/simplify gradient computations, we typically take the logarithm of the likelihood function, which gives
\begin{equation}
\begin{aligned}
\log p(T \vert X) &= \sum_{n=1}^{N} p(t_n \vert x_n) \\
\mathcal{L}(\boldsymbol\omega, \sigma^2) &= \sum_{n=1}^{N} \left( -\frac{1}{2}\log \left( 2\pi \right) -\frac{1}{2}\log \left( \sigma^2 \right) - \frac{1}{2\sigma^2} \left[ t_n - f(x_n)\right] ^2 \right), \\
\end{aligned}
\end{equation}

where $\mathcal{L}(\mathbf{w}, \sigma^2)$ is the log-likelihood function that we wish to maximise. To maximise this function, we can use numerous methods (grid search, optimisation), but let us proceed with an analytical procedure to deriving the estimators $g(\hat{\boldsymbol\omega})$ and $g(\hat{\sigma}^2)$. 

Let us derive $\mathcal{L}(\boldsymbol\omega, \sigma^2)$ with respect to $\boldsymbol\omega$:
\begin{equation}
\begin{aligned}
0 &= \frac{d\mathcal{L}}{d\boldsymbol\omega} \\
&= \frac{d\mathcal{L}}{d\mathcal{f}} \frac{df}{d\boldsymbol\omega} \\
&= \sum_{n=1}^{N}\frac{1}{\sigma^2} \left[ t_n - f(x_n) \right] \frac{df}{d\boldsymbol\omega}  \\
&= \sum_{n=1}^{N}\frac{1}{\sigma^2} \left[ t_n - f(x_n) \right]\boldsymbol\phi^T(x_n)\\
&= \sum_{n=1}^{N} t_n \boldsymbol\phi^T(x_n) - f(x_n) \boldsymbol\phi^T(x_n) \\
&= \sum_{n=1}^{N} t_n \boldsymbol\phi^T(x_n) - \boldsymbol\omega^t \boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n) \\
&= \sum_{n=1}^{N} t_n \boldsymbol\phi^T(x_n) - \boldsymbol\omega^t \sum_{n=1}^{N} \boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n)
\end{aligned}
\end{equation}
where the derivative $\frac{df}{d\boldsymbol\omega}$ is easy found using Equation (69) from the __[Matrix cookbook](https://www.math.uwaterloo.ca/~hwolkowi/matrixcookbook.pdf)__. Now, let us think about what the observed equation tells us. In the first term we are weighting  $\boldsymbol\phi^T(x_n)$ by $t_n$, and this can be easily expanded into a vector form through (where we can go through the visual idea)

\begin{equation}
\begin{bmatrix}
t_1  & t_2  & \vdots & t_N \\
\end{bmatrix}
\begin{bmatrix}
\phi_1^{(0)} & \phi_1^{(1)} & \cdots & \phi_1^{(K)} \\
\phi_2^{(0)} & \phi_2^{(1)} & \cdots & \phi_2^{(K)} \\
\vdots & \vdots & \ddots & \vdots \\
\phi_N^{(0)} & \phi_N^{(1)} & \cdots & \phi_N^{(K)} \\
\end{bmatrix}
= \mathbf{t}^T\boldsymbol\Phi,
\end{equation}

where 
\begin{equation}
\boldsymbol\Phi = 
\begin{bmatrix}
\phi_1^{(0)} & \phi_1^{(1)} & \cdots & \phi_1^{(K)} \\
\phi_2^{(0)} & \phi_2^{(1)} & \cdots & \phi_2^{(K)} \\
\vdots & \vdots & \ddots & \vdots \\
\phi_N^{(0)} & \phi_N^{(1)} & \cdots & \phi_N^{(K)} \\
\end{bmatrix},
\end{equation}
such that the $n^{th}$ row of $\boldsymbol\Phi$ is equal to $\boldsymbol\phi(x_n)^T$ and $\boldsymbol\Phi \in \mathbb{R}^{N \times K + 1}$. If we inspect the shape of the resulting matrix, we have a $\mathbb{R}^{(1 \times N) \times (N \times K + 1)} = \mathbb{R}^{1 \times K + 1}$ matrix, which we would expect based on the current form of the first term, which is a sum of $t_n$ weighted row vectors.

In term two of the objective function, we see that we have a summation over $N$ of the outer product $\boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n)$. Let us first expand this outer product and then think about methods to vectorise this process. The outer product $\boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n)$ is equal to
\begin{equation}
\begin{aligned}
\boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n)
&= \begin{bmatrix}
\phi^{(0)}(x_n) \\
\phi^{(1)}(x_n) \\
\vdots \\
\phi^{(K)}(x_n)
\end{bmatrix}
\begin{bmatrix}
\phi^{(0)}(x_n) & \phi^{(1)}(x_n) & \cdots & \phi^{(K)}(x_n)
\end{bmatrix} \\
&= 
\begin{bmatrix}
\phi^{(0)}(x_n)\phi^{(0)}(x_n) & \phi^{(0)}(x_n)\phi^{(1)}(x_n) & \cdots & \phi^{(0)}(x_n)\phi^{(K)}(x_n) \\
\phi^{(1)}(x_n)\phi^{(0)}(x_n) & \phi^{(1)}(x_n)\phi^{(1)}(x_n) & \cdots & \phi^{(1)}(x_n)\phi^{(K)}(x_n) \\
\vdots & \vdots & \ddots & \vdots \\
\phi^{(K)}(x_n)\phi^{(0)}(x_n) & \phi^{(K)}(x_n)\phi^{(1)}(x_n) & \cdots & \phi^{(K)}(x_n)\phi^{(K)}(x_n) \\
\end{bmatrix}.
\end{aligned}
\end{equation}

If we sum $N$ of these outer products, we end up with $N$ matrices of size $\mathbb{R}^{K \times K}$ that are summed together. Can we vectorise this? We can, and it looks like:
\begin{equation}
\sum_{n=1}^N \boldsymbol\phi(x_n) \boldsymbol\phi^T(x_n) = \boldsymbol\Phi^T \boldsymbol\Phi,
\end{equation}

if we perform a visual inspection of this, using the presented form of $\boldsymbol\Phi$, we see that this is fine.

\begin{equation}
\boldsymbol\Phi^T \boldsymbol\Phi = 
\begin{bmatrix}
\phi_1^{(0)} & \phi_2^{(0)} & \cdots & \phi_N^{(0)}  \\
\phi_1^{(1)} & \phi_2^{(1)} & \cdots & \phi_N^{(1)} \\
\vdots & \vdots & \ddots & \vdots \\
\phi_1^{(K)} & \phi_2^{(K)} & \cdots & \phi_N^{(K)}
\end{bmatrix}
\begin{bmatrix}
\phi_1^{(0)} & \phi_1^{(1)} & \cdots & \phi_1^{(K)} \\
\phi_2^{(0)} & \phi_2^{(1)} & \cdots & \phi_2^{(K)} \\
\vdots & \vdots & \ddots & \vdots \\
\phi_N^{(0)} & \phi_N^{(1)} & \cdots & \phi_N^{(K)} \\
\end{bmatrix},
\end{equation}

Thus, we have the following derivative, from which we can simplify terms to 
\begin{equation}
\begin{aligned}
0 &= \mathbf{t}^T\boldsymbol\Phi  - \boldsymbol\omega^T \boldsymbol\Phi^T \boldsymbol\Phi \\
\boldsymbol\omega^T \boldsymbol\Phi^T \boldsymbol\Phi &= \mathbf{t}^T\boldsymbol\Phi \\,
\end{aligned}
\end{equation}
which is almost a solvable linear system of equations in the form $\mathbf{A}\mathbf{x} = \mathbf{b}$, but we need to put in just a little bit more work to solve the final system. By using the transpose, we get
\begin{equation}
\begin{aligned}
\boldsymbol\Phi^T \boldsymbol\Phi \boldsymbol\omega  &= \boldsymbol\Phi^T \mathbf{t} \\
\hat{\boldsymbol\omega}  &= \left( \boldsymbol\Phi^T \boldsymbol\Phi \right)^{-1} \boldsymbol\Phi^T \mathbf{t},
\end{aligned}
\end{equation}
which gives us an analytical equation to solve for the maximum likelihood estimate of $\boldsymbol\omega$.

Finally, let us derive $\mathcal{L}(\boldsymbol\omega, \sigma^2)$ with respect to $\sigma^2$:
\begin{equation}
\begin{aligned}
0 &= \frac{d\mathcal{L}}{d\sigma^2} \\
&= \sum_{n=1}^{N} \left( -\frac{1}{2\sigma^2} + \frac{1}{2(\sigma^2)^2} \right)\left[ t_n - f(x_n)\right] ^2 \\
&= \frac{1}{2\sigma^2}\sum_{n=1}^{N} \left( -1 + \frac{1}{\sigma^2} \right)\left[ t_n - f(x_n)\right] ^2 \\
&= \frac{1}{2\sigma^2}\left(-N + \frac{1}{\sigma^2}\sum_{n=1}^{N} \left[ t_n - f(x_n)\right] ^2 \right) \\
\end{aligned}
\end{equation}
and, after some re-arranging we get
\begin{equation}
\hat{\sigma}^2 = \frac{1}{N}\sum_{n=1}^{N} \left[ t_n - f(x_n)\right] ^2
\end{equation}

Thus, we have two estimators of interest for which we want to find roots for 

$g(\hat{\boldsymbol\omega}) = \frac{d\mathcal{L}}{d\boldsymbol\omega}$ 

and 

$g(\hat{\sigma}^2) = \frac{d\mathcal{L}}{d\sigma^2}$

and we can obtain the roots to these estimators analytically using

$\hat{\boldsymbol\omega} = \left( \boldsymbol\Phi^T \boldsymbol\Phi \right)^{-1} \boldsymbol\Phi^T \mathbf{t}$

and 

$\hat{\sigma}^2 = \frac{1}{N}\sum_{n=1}^{N} \left[ t_n - f(x_n)\right] ^2$

### Find the maximum likelihood estimate of the unknown parameters for the given data.

```python
#Step 1 - define the Phi matrix function
def Phi_func(X, K):
    
    return X**np.arange(K + 1)

def maximum_likelihood_solution(X, t, K, biased = True, verbose = False):

    #calculate Phi
    Phi = Phi_func(X, K)

    #calculate ATA
    ATA = np.dot(Phi.T, Phi)

    #Calculate ATb
    ATb = np.dot(Phi.T, t)

    #Check shapes are K + 1 x K + 1 and K + 1 x 1
    #print(ATA.shape, ATb.shape)

    #Calculate Omega
    omega_ML = np.linalg.solve(ATA, ATb)

    #Calculate sigma
    t_pred = f_x(X, omega_ML)

    if biased:
        divisor = X.shape[0]
    
    else:
        divisor = X.shape[0] - K - 1
    
    sigma_ML = np.sqrt(1/divisor * np.sum((t_pred - t)**2, axis = 0))

    #Calculate the log-likelihood:
    LL = log_likelihood_function(X, t, omega_ML, sigma_ML)[0]

    if verbose: 
        print("For a {}th order polynomial:".format(K))
        print("\nOptimal weights:\n{}\n".format(omega_ML))
        print("Optimal sigma:\n{}\n".format(sigma_ML))
        print("Optimal model log-likelihood:\n{}\n".format(LL))
    
    return omega_ML, sigma_ML, LL

def confidence_interval(parameters, X, gamma = 0.9, weight_index = 0):
    
    #Define the student-t distribution
    pdf_t = scistats.t
    N = X.shape[0]
    K = parameters[0].shape[0] - 1
    print(N, K)
    dof = N - K - 1
    
    #Unbiased variance
    mean = parameters[0][weight_index]
    sigma_unbiased = parameters[1]**2 * (N)/(N - K - 1) #parameter[1] is the standard deviation
    
    Phi = Phi_func(X, K)
    
    A = np.linalg.inv(np.dot(Phi.T, Phi))
    
    Sigma = sigma_unbiased * A

    lower_bound = pdf_t.ppf(q = (1 - gamma)/2, df = dof, loc = mean, scale = np.sqrt(Sigma[weight_index, weight_index]))[0]
    upper_bound = pdf_t.ppf(q = (1 + gamma)/2, df = dof, loc = mean, scale = np.sqrt(Sigma[weight_index, weight_index]))[0]
    
    return lower_bound, upper_bound
    

#Visualise the optimal model as a function of the model order
k_range = np.arange(0, 10, 1)
x = np.linspace(np.min(x_data), np.max(x_data), 1000)
plt.figure(figsize = (12, 12))

plt.plot(x_data, y_data, color = "b", linestyle = "--", marker = "x", label = "data samples")

c = plt.cm.cool(k_range / k_range[-1])

lower_bounds = []
upper_bounds = []

for K in k_range:
    omega_ML, sigma_ML, LL = maximum_likelihood_solution(x_data.reshape(-1, 1),
                                                         y_data.reshape(-1, 1), 
                                                         K,
                                                         biased = True,
                                                         verbose = False)
    
    model_x = f_x(x.reshape(-1, 1), omega_ML)

    plt.plot(x, model_x, color = c[K], label = "{} order polynomial, LL = {:.3f}".format(K, LL))
    
    LB, UB = confidence_interval((omega_ML, sigma_ML), 
                                 x_data.reshape(-1, 1), 
                                 gamma = 0.9, 
                                 weight_index = 0)
    
    print("\nOrder = {}".format(K))
    print(np.round(omega_ML.T, 6))
    print(np.round(sigma_ML, 6))
    print(LL)
    
    lower_bounds.append(LB)
    upper_bounds.append(UB)
    
    
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("./images/Q2a_fig3.png")
plt.show()
```

### Calculate a two-sided 90% confidence interval on the parameter w0.

The $100 \cdot \gamma\%$ confidence intervals of the $k^{th}$ weight is given as
\begin{equation}
\hat{\omega}_k - t_{(1+\gamma)/2} \sqrt{\Sigma_{kk}} \leq \omega_k \leq \hat{\omega}_k + t_{(1+\gamma)/2} \sqrt{\Sigma_{kk}},
\end{equation}

where $t_{(1+\gamma)/2}$ is the $t$-statistic with $N - K - 1$ degrees of freedom. This can be re-written as
\begin{equation}
t_{(1-\gamma)/2}\left(\hat{\omega}_k, \Sigma_{kk}\right) \leq \omega_k \leq t_{(1+\gamma)/2}\left(\hat{\omega}_k, \Sigma_{kk}\right) 
\end{equation}

The covariance matrix $\boldsymbol\Sigma$ is given as
\begin{equation}
\boldsymbol\Sigma = \tilde{\sigma}^2 \mathbf{A},
\end{equation}

and the matrix $\mathbf{A}$ is equal to
\begin{equation}
\mathbf{A} = \left( \boldsymbol\Phi^T\boldsymbol\Phi \right)^{-1}.
\end{equation}

It is important to note here that $\tilde{\sigma}^2$ refers to the unbiased maximum likelihood estimate of the linear regression model variance, given as 
\begin{equation}
\tilde{\sigma}^2 = \frac{N}{N - K - 1}\hat{\sigma}^2.
\end{equation}

#### Define all important functions

```python
#Place all important functions here
def sampler(n_samples, omega, sigma, x_range = [0, 10]):
    
    x_n = np.random.uniform(x_range[0], x_range[1], n_samples).reshape(-1, 1)
    
    noise = np.random.randn(n_samples).reshape(-1, 1) * sigma
    
    y_n = f_x(x_n, omega) + noise
    
    return x_n, y_n

def feature_transform(X, K):
    
    return X**np.arange(K + 1)

def f_x(X, omega):
    
    K = omega.shape[0] - 1 #Index starts at zero
    
    Xtransform = feature_transform(X, K)
    
    return np.dot(Xtransform, omega)

def mean_squared_error(t, t_pred, sigma = None, n_params = None, n_samples = None, correction = False):#n_samples and correction do nothing here): #sigma is a useless parameter, just kept in for metric consistency

    return np.sum((t - t_pred)**2, axis = 0)[0]

def log_likelihood_function(t, t_pred, sigma, n_params = None, n_samples = None, correction = False):#n_samples and correction do nothing here
    
    term1 = -1/2 * np.log(2 * np.pi)
    term2 = -1/2 * np.log(sigma**2)
    
    term3 = -1/(2 * sigma**2) * mean_squared_error(t, t_pred)
    
    log_likelihood = term1 + term2 + term3
    
    return log_likelihood

def negative_log_likelihood_function(t, t_pred, sigma, n_params = None, n_samples = None, correction = False):#n_samples and correction do nothing here
    
    return -1 * log_likelihood_function(t, t_pred, sigma, n_params, n_samples, correction)

def aikes_information_criterion(t, t_pred, sigma, n_params, n_samples = None, correction = False):
    
    nll = negative_log_likelihood_function(t, t_pred, sigma)
    
    AIC = 2 * n_params + 2 * nll
    
    if correction:
        AIC += (2 * n_params**2 + 2 * n_params) / (n_samples - n_params - 1)
    
    return AIC

def bayesian_information_criterion(t, t_pred, sigma, n_params, n_samples, correction = False): #Correction does nothing here
    
    nll = negative_log_likelihood_function(t, t_pred, sigma, n_params, n_samples, correction)
    
    BIC = n_params * np.log(n_samples) + 2 * nll
    
    return BIC

def maximum_likelihood_solution(X, t, K, biased = True, verbose = False):

    #calculate Phi
    Phi = Phi_func(X, K)

    #calculate ATA
    ATA = np.dot(Phi.T, Phi)

    #Calculate ATb
    ATb = np.dot(Phi.T, t)

    #Check shapes are K + 1 x K + 1 and K + 1 x 1
    #print(ATA.shape, ATb.shape)

    #Calculate Omega
    omega_ML = np.linalg.solve(ATA, ATb)

    #Calculate sigma
    t_pred = f_x(X, omega_ML)

    if biased:
        divisor = X.shape[0]
    
    else:
        divisor = X.shape[0] - K - 1
    
    sigma_ML = np.sqrt(1/divisor * mean_squared_error(t, t_pred))#0.1#

    if verbose: 
        print("For a {}th order polynomial:".format(K))
        print("\nOptimal weights:\n{}\n".format(omega_ML))
        print("Optimal sigma:\n{}\n".format(sigma_ML))
    
    return omega_ML, sigma_ML

def standard_validation(X, t, k, normalise = True, correction = False, metric = mean_squared_error):
    
    n_samples = X.shape[0]
    
    #Train the model on X
    parameters = maximum_likelihood_solution(X, 
                                             t, 
                                             k, 
                                             biased = True, 
                                             verbose = False)
    
    t_pred = f_x(X, parameters[0])
    
    n_params = parameters[0].shape[0] + 1
    
    metric_value = metric(t, t_pred, parameters[1], n_params, n_samples, correction)
    
    if normalise:
            metric_value /= n_samples
    
    return metric_value
    

def cross_validation(X, t, k, n_folds, normalise = True, correction = False, metric = mean_squared_error):
    
    N = X.shape[0]
    fold_size = N // n_folds
    
    if int(fold_size * n_folds) != N:
        print("The number of folds is not equally divisible into N (n_folds = {}, N = {}). {} points will not be allocated to a batch.".format(n_folds, N, N - int(fold_size * n_folds)))
    
    #Split the data into n_fold blocks
    
    cv_error = np.zeros(n_folds)
    
    #Train and evaluate the data on the Xtrain and Xtest
    for i in range(n_folds):
        
        start = int(i * fold_size)
        end = int((i + 1) * fold_size)
        
        train_ix = list(range(0, start, 1)) + list(range(end, N, 1))
        test_ix = list(range(start, end, 1))
        
        X_train = X[train_ix, :]
        X_test = X[test_ix, :]
        
        t_train = t[train_ix, :]
        t_test = t[test_ix, :]
        
                       
        #Fit the model on X_train and t_train
        parameters = maximum_likelihood_solution(X_train, 
                                                 t_train, 
                                                 k, 
                                                 biased = True, 
                                                 verbose = False)
                       
        #Evaluate the data on X_test
        t_pred_test = f_x(X_test, parameters[0])
        
        n_params = parameters[0].shape[0] + 1
        n_samples = X_test.shape[0]
        
        #Store cv error
        cv_error[i] = metric(t_test, t_pred_test, parameters[1], n_params, n_samples, correction)
        
        if normalise:
            cv_error[i] /= t_pred_test.shape[0]
        
    
    #Calculate CV error average
    cv_avg = np.mean(cv_error)
    
    return cv_avg
```

How do we calculate these terms?

$variance(x) = \mathbb{E} \{ (x^{0}\cdot\omega_0 + x^{1}\cdot\omega_1 + \cdots + x^{K}\cdot\omega_K) - \mathbb{E}\{ x^{0}\cdot\omega_0 + x^{1}\cdot\omega_1 + \cdots + x^{K}\cdot\omega_K\})^2 \}$

#### Helper function

```python
def bias_variance_investigation(n_experiments, omega_optimal, sigma, n_samples, polynomial_orders, normalise_flag = True,  correction_flag = True):
    
    #Sampler parameters
    #omega_optimal = np.array([[3, 0.2, 0.7, -0.45, 0.2, -0.1]]).T #Term1 = DC offset
    #sigma = 0.1
    #n_samples = 200
    
    metric_name = ["Negative log-likelihood", "AIC", "BIC", "MSE"]
    metrics_used = [negative_log_likelihood_function, aikes_information_criterion, bayesian_information_criterion, mean_squared_error]

    
    x_comparison_range = np.linspace(-2, 3, 1000).reshape(-1, 1)
    
    f_x_optimal = f_x(x_comparison_range, omega_optimal) #The optimal model responses as a function of x
    
    #Reshape, expand and repeat
    f_x_optimal = np.repeat(f_x_optimal, repeats = len(metrics_used), axis = 1)
    
    bias_expectation_sv = np.zeros((1000, len(metrics_used))) #N x n_metrics vector
    bias_expectation_cv = np.zeros((1000, len(metrics_used))) #N x n_metrics vector
    
    #Determine bias expectation by repeating the experiments
    for experiment_no in range(n_experiments):
        
        #For the given range of omegas, find the optimal model order from the different methods
        #and then evaluate model over x_comparison_range, store it in bias_expectation
        #and then at the end, normalise by the number of experiments.
        
        #Sample the data
        X, t = sampler(n_samples, 
                       omega_optimal, 
                       sigma, 
                       x_range = [-2, 3])

        metric_matrix_sv = np.zeros((len(polynomial_orders), len(metrics_used)))
        metric_matrix_cv = np.zeros((len(polynomial_orders), len(metrics_used)))

        for k_iter in polynomial_orders:

            k_metrics_sv = []
            k_metrics_cv = []

            for metrics in metrics_used:

                #Standard validation
                metric_val_sv = standard_validation(X, 
                                                     t, 
                                                     k_iter,
                                                     normalise = normalise_flag, 
                                                     correction = correction_flag, 
                                                     metric = metrics)

                k_metrics_sv.append(metric_val_sv)

                #Cross-validation
                metric_val_cv = cross_validation(X, 
                                                  t, 
                                                  k_iter, 
                                                  n_folds = 10,
                                                  normalise = normalise_flag,
                                                  correction = correction_flag, 
                                                  metric = metrics)

                k_metrics_cv.append(metric_val_cv)


            metric_matrix_sv[k_iter, :] = k_metrics_sv
            metric_matrix_cv[k_iter, :] = k_metrics_cv
        
        #Find the optimal model orders for each metric
        sv_optimal_orders = np.argmin(metric_matrix_sv, axis = 0)
        cv_optimal_orders = np.argmin(metric_matrix_cv, axis = 0)
        
        print(sv_optimal_orders, cv_optimal_orders)

        #Fit the model for each case and update the bias calculation
        
        for i in range(len(sv_optimal_orders)):
            parameters_sv = maximum_likelihood_solution(X, 
                                                         t, 
                                                         sv_optimal_orders[i], 
                                                         biased = True, 
                                                         verbose = False)
            
            parameters_cv = maximum_likelihood_solution(X, 
                                                         t, 
                                                         cv_optimal_orders[i], 
                                                         biased = True, 
                                                         verbose = False)
            
            bias_expectation_sv[:, i] += f_x(x_comparison_range, parameters_sv[0])[:, 0]
            bias_expectation_cv[:, i] += f_x(x_comparison_range, parameters_cv[0])[:, 0]
    
    bias_expectation_sv /= n_experiments
    bias_expectation_cv /= n_experiments
    
    bias_sv = f_x_optimal - bias_expectation_sv
    bias_cv = f_x_optimal - bias_expectation_cv
    
    bias = (bias_sv, bias_cv)
    
    #Determine the variance
    variance_sv = np.zeros_like(bias_expectation_cv)
    variance_cv = np.zeros_like(bias_expectation_cv)
    
    for experiment_no in range(n_experiments):
        
        #For the given range of omegas, find the optimal model order from the different methods
        #and then evaluate model over x_comparison_range, store it in bias_expectation
        #and then at the end, normalise by the number of experiments.
        
        #Sample the data
        X, t = sampler(n_samples, 
                       omega_optimal, 
                       sigma, 
                       x_range = [-2, 2])

        metric_matrix_sv = np.zeros((len(polynomial_orders), len(metrics_used)))
        metric_matrix_cv = np.zeros((len(polynomial_orders), len(metrics_used)))

        for k_iter in polynomial_orders:

            k_metrics_sv = []
            k_metrics_cv = []

            for metrics in metrics_used:

                #Standard validation
                metric_val_sv = standard_validation(X, 
                                                     t, 
                                                     k_iter,
                                                     normalise = normalise_flag, 
                                                     correction = correction_flag, 
                                                     metric = metrics)

                k_metrics_sv.append(metric_val_sv)

                #Cross-validation
                metric_val_cv = cross_validation(X, 
                                                  t, 
                                                  k_iter, 
                                                  n_folds = 10,
                                                  normalise = normalise_flag,
                                                  correction = correction_flag, 
                                                  metric = metrics)

                k_metrics_cv.append(metric_val_cv)


            metric_matrix_sv[k_iter, :] = k_metrics_sv
            metric_matrix_cv[k_iter, :] = k_metrics_cv
        
        #Find the optimal model orders for each metric
        sv_optimal_orders = np.argmin(metric_matrix_sv, axis = 0)
        cv_optimal_orders = np.argmin(metric_matrix_cv, axis = 0)

        #Fit the model for each case and update the bias calculation
        
        for i in range(len(sv_optimal_orders)):
            parameters_sv = maximum_likelihood_solution(X, 
                                                         t, 
                                                         sv_optimal_orders[i], 
                                                         biased = True, 
                                                         verbose = False)
            
            parameters_cv = maximum_likelihood_solution(X, 
                                                         t, 
                                                         cv_optimal_orders[i], 
                                                         biased = True, 
                                                         verbose = False)
            
            variance_sv[:, i] += (f_x(x_comparison_range, parameters_sv[0])[:, 0] - bias_expectation_sv[:, i])**2
            variance_cv[:, i] += (f_x(x_comparison_range, parameters_cv[0])[:, 0] - bias_expectation_cv[:, i])**2
    
    variance_sv /= n_experiments
    variance_cv /= n_experiments
    
    variance = (variance_sv, variance_cv)
    
        
    return x_comparison_range, bias, variance
```

### Formulate the maximum likelihood estimation problem of the two unknown parameters $\boldsymbol\theta = \left[\theta_1, \theta_2\right]$ by assuming the noise is Gaussian distributed. Find the maximum likelihood estimates for the generated data. Superimpose the function $\tilde{\theta_1}\cdot \sin \left( 2 \cdot \pi \cdot t \right)$ over the data, where the maximum likelihood estimate is denoted as $\tilde{\theta_1}$.

Process very similar to Q2.2 and is as follows: Given a dataset $\mathcal{D} = \{ \mathbf{T}, \mathbf{x}\}$, where $\mathbf{T} = [t_1, \cdots, t_N]^T$ and $\mathbf{x} = [x_1, \cdots, x_N]^T$, $\mathbf{T} \in \mathbb{R}^{N \times 1}$, $\mathbf{x} \in \mathbb{R}^{N \times 1}$. We assume that the following model:

\begin{equation}
x_n = f(t_n; \theta_1) + \epsilon_n,
\end{equation}

where $\epsilon_n \sim \mathcal{N}(0, \theta_2)$ and $\boldsymbol\theta = \left[\theta_1, \theta_2\right]$. let $f(t_n; \theta_1)$ be

\begin{equation}
f(t; \theta_1) = \theta_1 \sin \left(2 \cdot \pi \cdot t \right).
\end{equation}

The resulting conditional distribution is
\begin{equation}
p(x \vert t, \boldsymbol\theta) = \mathcal{N}(x \vert f(t), \theta_2^2) = \frac{1}{\sqrt{2\pi\theta_2^2}} \exp\left( -\frac{1}{2\theta_2^2}\left[x_n - f(t_n;\theta_1) \right]^2\right)
\end{equation}



Now I will not go into detail about the log-likelihood function, but we expect the function to be in the form:
\begin{equation}
\begin{aligned}
\log p(\mathbf{x} \vert \mathbf{T}) &= \sum_{n=1}^{N} \log  p(x_n \vert t_n) \\
\mathcal{L}(\boldsymbol\theta) &= \sum_{n=1}^{N} \left( -\frac{1}{2}\log \left( 2\pi \right) -\frac{1}{2}\log \left( \theta_2^2 \right) - \frac{1}{2\theta_2^2} \left[ x_n - f(t_n;\theta_1)\right] ^2 \right), \\
\end{aligned}
\end{equation}

Let us derive the estimators for this problem.

\begin{equation}
\begin{aligned}
0 &= \frac{d\mathcal{L}}{d\theta_1} \\
&= \frac{d\mathcal{L}}{df} \cdot \frac{df}{d\theta_1} \\
&= 0 - 0 -\frac{1}{2\theta_2^2} \sum_{n=1}^{N} \left[x_n - f(t_n)\right]\cdot \sin \left(2 \cdot \pi \cdot t_n \right) \\
\sum_{n=1}^{N} \theta_1 \sin \left(2 \cdot \pi \cdot t_n \right)\cdot \sin \left(2 \cdot \pi \cdot t_n \right) &=  \sum_{n=1}^{N} x_n \sin \left(2 \cdot \pi \cdot t_n \right) \\
\theta_1 &= \frac{\sum_{n=1}^{N} x_n \sin \left(2 \cdot \pi \cdot t_n \right)}{\sum_{n=1}^{N} \sin \left(2 \cdot \pi \cdot t_n \right)\cdot \sin \left(2 \cdot \pi \cdot t_n \right)}
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
0 &= \frac{d\mathcal{L}}{d\theta_2} \\
&= \sum_{n=1}^{N} \left( 0 -\frac{2\theta_2}{2\theta_2^2} -\frac{-2}{2\theta_2^3}\left[ x_n - f(t_n;\theta_1)\right]^2 \right) \\
&= \sum_{n=1}^{N} \left( 0 - \frac{1}{\theta_2} + \frac{1}{\theta_2^3}\left[ x_n - f(t_n;\theta_1)\right]^2  \right) \\
&= \frac{1}{\theta_2} \sum_{n=1}^{N} \left( - 1 + \frac{1}{\theta_2^2}\left[ x_n - f(t_n;\theta_1)\right]^2  \right) \\
N \theta_2^2 &= \sum_{n=1}^{N}\left[ x_n - f(t_n;\theta_1)\right]^2 \\
\theta_2 &= \sqrt{\frac{1}{N}\sum_{n=1}^{N}\left[ x_n - f(t_n;\theta_1)\right]^2} \\
\end{aligned}
\end{equation}

```python
%matplotlib inline
#Determine theta_1
numerator = np.sum(x_data * np.sin(2 * np.pi * t_data))
denominator = np.sum(np.sin(2 * np.pi * t_data) * np.sin(2 * np.pi * t_data))
theta_1_ML = numerator / denominator

#determine theta_2
theta_2_ML = np.sqrt(np.mean((x_data - f_x(t_data, [theta_1_ML, None]))**2))
theta_ML_gauss_analytical = [theta_1_ML, theta_2_ML]

print(theta_ML_gauss_analytical)

gauss_std = theta_ML_gauss_analytical[1]

gauss_mu = f_x(t, theta_ML_gauss_analytical)
gauss_lower_range = gauss_mu - 3 * gauss_std
gauss_upper_range = gauss_mu + 3 * gauss_std

plt.figure(figsize = (12, 12))
plt.title(r"Analytical Gaussian maximum likelihood solution: $\left[\theta_1, \theta_2\right] =  \left[{:.4f}, {:.4f}\right]$".format(theta_ML_gauss_analytical[0], theta_ML_gauss_analytical[1]))
plt.scatter(t_data, x_data, color ="b", marker = "x", label = "Observed data")
plt.plot(t, gauss_mu, label = r'$\mu = f(t, \theta_1)$')
plt.fill_between(t, gauss_lower_range, gauss_upper_range, alpha = 0.5, color = "r", label = r"$\theta_1 \pm 3 \theta_2$")
plt.grid()
plt.legend()
plt.xlabel("t")
plt.ylabel("x(t)")
plt.savefig("./images/Q3b_fig2.png")
plt.show()
```

### Formulate the maximum likelihood estimation problem of the two unknown parameters by assuming that the noise is Laplacian distributed.

Process very similar to Q2.2 and is as follows: Given a dataset $\mathcal{D} = \{ \mathbf{T}, \mathbf{x}\}$, where $\mathbf{T} = [t_1, \cdots, t_N]^T$ and $\mathbf{x} = [x_1, \cdots, x_N]^T$, $\mathbf{T} \in \mathbb{R}^{N \times 1}$, $\mathbf{x} \in \mathbb{R}^{N \times 1}$. We assume that the following model:

\begin{equation}
x_n = f(t_n; \theta_1) + \epsilon_n,
\end{equation}

where $\epsilon_n \sim L(0, \theta_2)$ and $\boldsymbol\theta = \left[\theta_1, \theta_2\right]$. let $f(t_n; \theta_1)$ be

\begin{equation}
f(t; \theta_1) = \theta_1 \sin \left(2 \cdot \pi \cdot t \right).
\end{equation}

The resulting conditional distribution is
\begin{equation}
p(x \vert t, \boldsymbol\theta) = L(x \vert f(t), \theta_2) = \frac{1}{2\theta_2} \exp\left( -\frac{1}{2\theta_2}\vert x_n - f(t_n;\theta_1) \vert \right)
\end{equation}



Now I will not go into detail about the log-likelihood function, but we expect the function to be in the form:
\begin{equation}
\begin{aligned}
\log p(\mathbf{x} \vert \mathbf{T}) &= \sum_{n=1}^{N} \log p(x_n \vert t_n) \\
\mathcal{L}(\boldsymbol\theta) &= \sum_{n=1}^{N} \left( -\log(2\theta_2) -\frac{1}{\theta_2}\vert x - f(t_n;\theta_1) \vert \right), \\
\end{aligned}
\end{equation}

Let us derive the estimators for this problem.

\begin{equation}
\begin{aligned}
0 &= \frac{d\mathcal{L}}{d\theta_1} \\
&= \frac{d\mathcal{L}}{df} \cdot \frac{df}{d\theta_1} \\
\end{aligned}
\end{equation}

The complexity of this derivative comes from the absolute value over $\vert x - f \vert$. The derivative of an absolute value of an arbitrary function $f(x)$ is (let $u = f(x)$):
$\frac{d}{dx} \vert f(x) \vert = \frac{d}{du} \vert u \vert  \cdot \frac{du}{dx}= = \frac{d}{du} \vert u \vert  \cdot \frac{du}{dx} = \frac{u}{\vert u \vert} \cdot \frac{du}{dx} = \frac{f(x)}{\vert f(x) \vert} \frac{df}{dx} = \text{sgn}(f(x)) \frac{df}{dx}$

Thus, let $u = x_n - f(t_n; \theta_1)$:
\begin{equation}
\begin{aligned}
0 &=- \frac{1}{\theta_2}\sum_{n=1}^{N} \frac{d}{du} \vert u \vert  \cdot \frac{du}{d\theta_1}  \\
&= - \frac{1}{\theta_2}\sum_{n=1}^{N} \frac{u}{\vert u \vert}   \cdot \frac{du}{d\theta_1} \\
&= - \frac{1}{\theta_2}\sum_{n=1}^{N} \frac{x_n - f(t_n; \theta_1)}{\vert x_n - f(t_n; \theta_1) \vert}   \cdot -\sin \left(2 \cdot \pi \cdot t_n \right) \\
&= \frac{1}{\theta_2}\sum_{n=1}^{N} \frac{\sin \left(2 \cdot \pi \cdot t_n \right)\left(x_n - \theta_1 \sin \left(2 \cdot \pi \cdot t_n \right) \right)}{\vert x_n - \theta_1 \sin \left(2 \cdot \pi \cdot t_n \right) \vert} \\
\theta_1 \sum_{n=1}^{N}\frac{\sin \left(2 \cdot \pi \cdot t_n \right)\left( \sin \left(2 \cdot \pi \cdot t_n \right) \right) }{\vert x_n - \theta_1 \sin \left(2 \cdot \pi \cdot t_n \right) \vert} &= \sum_{n=1}^{N}\frac{\sin \left(2 \cdot \pi \cdot t_n \right)\left(x_n \right)}{\vert x_n - \theta_1 \sin \left(2 \cdot \pi \cdot t_n \right) \vert} \\
\end{aligned}
\end{equation}

It is clear that there is no analytical solution here as $\theta_1$ cannot be isolated. Thus, there is no closed-form solution to obtain the maximum likelihood estimate. Thus, we will turn to numerical optimisation methods.

```python
f_x = lambda t, theta: theta[0] * np.sin(2 * np.pi * t)

def log_gauss(x, mu, sigma):
    
    term1 = -1/2 * np.log(2 * np.pi)
    term2 = -1/2 * np.log(sigma**2)
    term3 = -1/(2 * sigma**2) * (x - mu)**2
    
    return term1 + term2 + term3

def log_laplace(x, mu, b):
    
    term1 = -1 * np.log(2 * np.abs(b) + 1e-12)
    term2 = -(1/b) * np.abs(x - mu)
    
    return term1 + term2

def log_likelihood_function(theta, t, x, log_func = log_gauss):
    
    x_pred = f_x(t, theta)
    
    ll_sum = np.sum(log_func(x, x_pred, theta[1]))
    
    return -1 * ll_sum

#Optimise Gaussian parameters
x0 = [1, 1]
opt_dict = sciopt.minimize(log_likelihood_function,
                          x0,
                          args = (t_data, x_data, log_gauss))

print(opt_dict)
theta_ML_gauss_sciopt = opt_dict['x']


#Optimise Laplacian parameters
x0 = [1, 1] #Use Gaussian parameters as the starting point
opt_dict = sciopt.minimize(log_likelihood_function,
                          x0,
                          args = (t_data, x_data, log_laplace))

print(opt_dict)
theta_ML_laplace_sciopt = opt_dict['x']

print("\n\nAnalytical ML Gaussian parameters:", theta_ML_gauss_analytical)
print("Scipy.optimize ML Gaussian parameters:", theta_ML_gauss_sciopt)
print("Scipy.optimize ML Laplacian parameters:", theta_ML_laplace_sciopt, "\n")

gauss_mu = f_x(t, theta_ML_gauss_sciopt)
gauss_std = theta_ML_gauss_sciopt[1]
gauss_lower_range = gauss_mu - 3 * gauss_std
gauss_upper_range = gauss_mu + 3 * gauss_std

laplace_mu = f_x(t, theta_ML_laplace_sciopt)
laplace_std = np.sqrt(2 * theta_ML_laplace_sciopt[1]**2)
laplace_lower_range = laplace_mu - 3 * laplace_std
laplace_upper_range = laplace_mu + 3 * laplace_std

plt.figure(figsize = (12, 12))
plt.title(r"Gaussian maximum likelihood solution: $\left[\theta_1, \theta_2\right] =  \left[{:.4f}, {:.4f}\right]$".format(theta_ML_gauss_sciopt[0], theta_ML_gauss_sciopt[1]))
plt.scatter(t_data, x_data, color ="b", marker = "x", label = "Observed data")
plt.plot(t, gauss_mu, label = r'$\mu = f(t, \theta_1)$')
plt.fill_between(t, gauss_lower_range, gauss_upper_range, alpha = 0.5, color = "r", label = r"$\theta_1 \pm 3 \theta_2$")
plt.grid()
plt.legend()
plt.xlabel("t")
plt.ylabel("x(t)")
plt.savefig("./images/Q3b_fig5.png")
plt.show()

plt.figure(figsize = (12, 12))
plt.title(r"Laplacian maximum likelihood solution: $\left[\theta_1, \theta_2\right] =  \left[{:.4f}, {:.4f}\right]$".format(theta_ML_laplace_sciopt[0], theta_ML_laplace_sciopt[1]))
plt.scatter(t_data, x_data, color ="b", marker = "x", label = "Observed data")
plt.plot(t, laplace_mu, label = r'$\mu = f(t, \theta_1)$')
plt.fill_between(t, laplace_lower_range, laplace_upper_range, alpha = 0.5, color = "r", label = r"$\theta_1 \pm 3 \sqrt{2\theta_2^2}$")
plt.grid()
plt.legend()
plt.xlabel("t")
plt.ylabel("x(t)")
plt.savefig("./images/Q3b_fig6.png")
plt.show()
```

For $\theta_1 = 5$ and $\theta_2 = 1$, generate data from a student-t with two degrees of freedom. Compare the performance of the maximum likelihood estimators of the Gaussian and Laplacian models (e.g., compare the bias and variance of $\tilde{\theta_1}$). Discuss the results.

```python
f_x = lambda t, theta: theta[0] * np.sin(2 * np.pi * t)

def student_t_generator(theta_1, theta_2, df, t_range = [0, 10], Fs = 100):
    
    t = np.arange(t_range[0], t_range[1], 1/Fs)
    noise_model = scistats.t(df = 2, loc = 0, scale = theta_2)
    
    #Create x data
    x_mean = f_x(t, [theta_1, theta_2])
    noise_samples = noise_model.rvs(len(t))
    x_samples = x_mean + noise_samples
    
    return t, x_samples

def parameter_bias_variance(theta_params, n_repeats, log_model, parameter_index = 0):
    #Assume here that we do investigate the variance around the bias, but rather just the bias and variance itself.
    
    #Hidden components
    n_samples_iter = 10**np.arange(1, 6, 1)
    
    parameter_expectation = np.zeros(len(n_samples_iter))
    bias = np.zeros(len(n_samples_iter))
    var = np.zeros(len(n_samples_iter))
    
    #Estimate the bias as a function of n_samples
    for cnt, n_samples in enumerate(n_samples_iter):
        
        param_list = []
        
        for i in range(n_repeats):
            
            #Sample the data
            t_data_sample, x_data_sample = student_t_generator(theta_params[0], 
                                                               theta_params[1], 
                                                               df = 2, 
                                                               t_range = [0, 10], 
                                                               Fs = n_samples / 10) #Divide by 10 as t_end * Fs = n_samples (re-arrange this to get Fs)

            x0 = [1, 1]
            opt_dict = sciopt.minimize(log_likelihood_function,
                                      x0,
                                      args = (t_data_sample, x_data_sample, log_model))

            theta_ML_repeats = opt_dict['x']
            
            param_list.append(theta_ML_repeats[parameter_index])
        
        #Store the expected bias and the variance around the bias
        parameter_expectation[cnt] = np.mean(param_list)
    
    bias = theta_params[parameter_index] - parameter_expectation
    
    #Estimate the variance as a function of n_samples
    for cnt, n_samples in enumerate(n_samples_iter):
        
        var_list = []
        
        for i in range(n_repeats):
            
            #Sample the data
            t_data_sample, x_data_sample = student_t_generator(theta_params[0], 
                                                               theta_params[1], 
                                                               df = 2, 
                                                               t_range = [0, 10], 
                                                               Fs = n_samples / 10) #Divide by 10 as t_end * Fs = n_samples (re-arrange this to get Fs)

            x0 = [1, 1]
            opt_dict = sciopt.minimize(log_likelihood_function,
                                      x0,
                                      args = (t_data_sample, x_data_sample, log_model))

            theta_ML_repeats = opt_dict['x']
            
            var_local = (theta_ML_repeats[parameter_index] - parameter_expectation)**2
            var_list.append(var_local)
        
        #Store the expected bias and the variance around the bias
        var[cnt] = np.mean(var_list)
    
    return bias, var, parameter_expectation
```

## Find the displacement and velocities of the unforced system, i.e., $f(t) = 0$ and plot the response as a function of time. Use a numerical integration method.

The system of interest linear dynamical system where we know the second order differential equation of the system, and we can write as a first order differential equation through

\begin{equation}
\begin{aligned}
\begin{bmatrix}
\dot{x} \\
\ddot{x}
\end{bmatrix} &= 
\begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m} \\
\end{bmatrix}
\begin{bmatrix}
x \\
\dot{x}
\end{bmatrix}+
\begin{bmatrix}
0 \\
\frac{f(t)}{m}
\end{bmatrix}, \\
&=
\begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m} \\
\end{bmatrix}
\begin{bmatrix}
x \\
\dot{x}
\end{bmatrix}+
\begin{bmatrix}
0 \\
\frac{1}{m}
\end{bmatrix}
\begin{bmatrix}
f(t)
\end{bmatrix}.
\end{aligned}
\end{equation}

In matrix-vector notation, this is given as

\begin{equation}
\dot{\mathbf{x}}(t) = \mathbf{A}\mathbf{x}(t) + \mathbf{B}\mathbf{u}(t)
\end{equation}

There are numerous methods available to solve of the ordinary differential equation if we know the initial state $\mathbf{x}(0)$ of the system. The problem that we seek a solution to is the computation the integral
\begin{equation}
x(t) = \int_{\mathbf{x}(0)}^{t}\dot{\mathbf{x}}(\tau)d\tau.
\end{equation}

however it may be difficult to perform this integration analytically. In this case, we can use numerical integration techniques for ordinary differential equations. The simplest of these is Euler's method, which replaces the derivative $\dot{\mathbf{x}}(t)$ with the forward finite difference approximation $\dot{\mathbf{x}}(t) \approx \frac{\mathbf{x}(t + \Delta t) - \mathbf{x}(t)}{\Delta t}$, where $\Delta t$ is referred to as the step size. We can re-arrange to develop an approximation for $\mathbf{x}(t + \Delta t)$ through
\begin{equation}
\begin{aligned}
\mathbf{x}(t + \Delta t) &\approx \mathbf{x}(t) + \Delta t \dot{\mathbf{x}}(t) \\
&\approx \mathbf{x}(t) + \Delta t  \left(  \mathbf{A}\mathbf{x}(t) + \mathbf{B}\mathbf{u}(t) \right) \\
&\approx \left( \Delta t \mathbf{A} + \mathbf{I} \right) \mathbf{x}(t) + \Delta t \mathbf{B}\mathbf{u}(t) \\
&\approx \mathbf{A}_d  \mathbf{x}(t) + \mathbf{B}_d \mathbf{u}(t), \\
\end{aligned}
\end{equation}

where
\begin{equation}
\mathbf{A}_d = \Delta t \mathbf{A} + \mathbf{I},
\end{equation}
and
\begin{equation}
\mathbf{B}_d = \Delta t \mathbf{B}.
\end{equation}

It is clear that we know have an approximate for the forward time state $\mathbf{x}(t + \Delta t)$, and we can use a recursive methodology to iteratively solve for $\mathbf{x}(t)$ on the interval $t \in \left[ 0, t_{end} \right]$, where $t_{end} >0$. The choice of the step size $\Delta t$ induces an error in referred to as the local truncation error, however information of this error is beyond the scope of this assignment. Interested readers can consult the works of [CITE] for more information. Thus, we can develop a discrete solution to the system using
\begin{equation}
\mathbf{x}(t_{k + 1}) = \mathbf{A}_d  \mathbf{x}(t_k) + \mathbf{B}_d \mathbf{u}(t_k),
\end{equation}
where $k \geq 0$ and $\mathbf{x}(0)$ must be given to initialise the solution.

Naturally, Euler's method is one of many numerical integration techniques for ordinary differential equations, but alternative methods will not be discussed here. Interested readers can refer to [CITE].

```python
#System properties
m = 1
c = 10
k = 1000
t_end = 10
Fs = 10000
t = np.arange(0, t_end, 1/Fs)

#Initial conditions
x0 = 1e-3
v0 = -1

#Forcing function
def forcing_function(t):
    
    return 0#100 * np.cos(2 * np.pi * t)

#First order DE matrices
A_mat = np.array([[0, 1],
                 [-k/m, -c/m]])

B_mat = np.array([[0], [1 / m]])

A_d = lambda dt, t: dt * A_mat + np.eye(A_mat.shape[0])
B_d = lambda dt, t: dt * B_mat
Ut = lambda t: np.array([[forcing_function(t)]])

def first_order_DE(t, x):

    return np.dot(A_mat, x) + np.dot(B_mat, Ut(t))

def discrete_DE(x_k, dt, t):
    
    x_new = np.dot(A_d(dt, t), x_k) + np.dot(B_d(dt, t), Ut(t))
    
    return x_new

def Euler(x0, dt, t):
    
    x0 = x0.reshape(-1, 1)
    
    Euler_solution = np.zeros((x0.shape[0], len(t)))
    Euler_solution[:, 0] = x0[:, 0]
    
    for cnt, t_value in enumerate(t[1:], start = 1):
        
        Euler_solution[:, cnt] = discrete_DE(Euler_solution[:, [cnt - 1]], dt, t_value)[:, 0]
    
    return Euler_solution
        

#Initial conditions
initial_conditions = np.array([x0, v0]) # [displacement(t=0) = 0, velocity(t=0) = 0]

#Euler solution
x_E = Euler(initial_conditions, 1/Fs, t)

#Runge-Kutta solution
solved_system_RK = solve_ivp(first_order_DE,
                          t_span = [0, t_end],
                          y0 = initial_conditions,
                          method ="RK45",
                          t_eval = t,
                          vectorized = True)

x_RK = solved_system_RK.y

#visualise
fig, ax = plt.subplots(1, 2, figsize = (10, 6))
ax[0].plot(t, x_E[0, :], "b-", label = "Euler")
ax[0].plot(t, x_RK[0, :], "r--", label = "Runge-kutta")
ax[0].legend()
ax[0].set_xlabel("Time (s)")
ax[0].set_ylabel(r"$x(t)$")
ax[0].set_title("Displacement")

ax[1].plot(t, x_E[1, :], "b-", label = "Euler")
ax[1].plot(t, x_RK[1, :], "r--", label = "Runge-kutta")
ax[1].legend()
ax[1].set_xlabel("Time (s)")
ax[1].set_ylabel(r"$\dot{x}(t)$")
ax[1].set_title("Velocity")

fig.tight_layout()
plt.savefig("./images/Q4a_fig1.png")
plt.show()
```

## Can you recover the original mass, damping and stiffness parameters from the system matrix? Discuss.

To recover the original parameters, we need some intrinsic knowledge of how the estimated system matrix is related to the true system matrix. If we assume that we know this relationship, for example (as was done in Q6.1.1 and Q6.1.2) if we know the governing physics of the problem (by using Newton's second law) to derive the ordinary differential equation (which contains the continuous system matrix), and by using finite differences to determine recursive equations used to find the discrete form of $\mathbf{x}(t)$ (which contains the discrete system matrix), then we at least have a reasonable starting point to try and recover these parameters. It is important to note that all of these components require a prior assumption regarding the physics of the data that we are modelling. If not, then there is definitely no way of doing so. 

If we assume that we know the relationship between the system matrix $\hat{\mathbf{A}}$ and the parameters of the model, then we can develop the following equation that relates the discrete-time system matrix $\hat{\mathbf{A}}_d$ to the continuous system matrix $\mathbf{A}_c$:
\begin{equation}
\begin{aligned}
\hat{\mathbf{A}}_d = \Delta t \mathbf{A}_c + \mathbf{I} \\
\mathbf{A}_c = \frac{\hat{\mathbf{A}}_d - \mathbf{I}}{\Delta t}.
\end{aligned}
\end{equation}

From this, we can try to develop the final relationship between the model parameters (which are found in the continuous system matrix $\mathbf{A}$) using the fact that we have $d$ degrees of freedom:
\begin{equation}
\begin{aligned}
\frac{\hat{\mathbf{A}}_d - \mathbf{I}}{\Delta t} =\mathbf{A}_c &= 
\begin{bmatrix}
\mathbf{0} & \mathbf{I} \\
-\mathbf{M}^{-1}\mathbf{K} & -\mathbf{M}^{-1}\mathbf{C} \\
\end{bmatrix} \\
\begin{bmatrix}
\mathbf{A}_{c, [1:dof, 1:2\cdot dof]} \\
\mathbf{A}_{c, [dof:2 \cdot dof, 1:2\cdot dof]}
\end{bmatrix}
&= \begin{bmatrix}
\mathbf{0} & \mathbf{I} \\
-\mathbf{M}^{-1}\mathbf{K} & -\mathbf{M}^{-1}\mathbf{C} \\
\end{bmatrix} \\
\begin{bmatrix}
\mathbf{A}_{c, [1:dof, 1:2\cdot dof]} \\
\mathbf{M} \mathbf{A}_{c, [dof:2 \cdot dof, 1:2\cdot dof]}
\end{bmatrix}
&= \begin{bmatrix}
\mathbf{0} & \mathbf{I} \\
-\mathbf{K} & -\mathbf{C} \\
\end{bmatrix} \\
\end{aligned}
\end{equation}

However, there is now a clear issue with the process as the left hand side of the second component $\mathbf{A}$ is a function of the system mass matrix $\mathbf{M}$. This complicates the solution as to recover any of the parameters we require knowledge of the masses of the system. Thus, I do not believe that one can recover all of the parameters, and you will need to have prior knowledge of one of the three components ($\mathbf{M}, \mathbf{C}, \mathbf{K}$) in order to be able to effectively solve for the other components.

If you do know one of the components, then we can decompose either the left or right hand size of Equation [CITE] to solve for the other parameters.

## Can you recover the original mass, damping and stiffness parameters from the system matrix? Discuss.

In this case, I believe that it is possible to recover the original system parameters. This is because the form of $\mathbf{B}_d$ allows us to recover the mass matrix exactly. If we assume that we know the governing physics of the system and the relationship between the continous system matrices $\mathbf{A}$ and $\mathbf{B}$ and the discrete system matrix $\hat{\mathbf{C}} = \left[ \mathbf{A}_d, \mathbf{B}_d \right]$, we obtain
\begin{equation}
\begin{aligned}
\mathbf{C} &= 
\begin{bmatrix} 
\mathbf{A}_d, & \mathbf{B}_d \\
\end{bmatrix} \\
&= 
\begin{bmatrix} 
\Delta t \mathbf{A} + \mathbf{I}, & \Delta t \mathbf{B} \\
\end{bmatrix} \\
&= 
\begin{bmatrix} 
\Delta t \mathbf{M}^{-1}
\begin{bmatrix}
\mathbf{0} & \mathbf{M} \\
-\mathbf{K} & -\mathbf{C} \\
\end{bmatrix} + \mathbf{I}, & \Delta t \begin{bmatrix} \mathbf{0} \\ \mathbf{M}^{-1} \end{bmatrix} \\
\end{bmatrix}
\end{aligned}
\end{equation}

Thus, we can decompose $\mathbf{C}$ to help us recover specific terms. Specifically, let us assume that there are $d$ degrees of freedom in the problem. If we adopt this notation, then the size of $\mathbf{C}$ is $\mathbb{R}^{\left( 2\cdot d \right) \times \left( 3 \cdot d \right)}$. Thus, we can decompose $\hat{\mathbf{C}}$ into
\begin{equation}
\hat{\mathbf{C}} = \begin{bmatrix} C_{[1:2d, 1:2d]} &  C_{[1:2d, 2d:3d]} \end{bmatrix}
= \begin{bmatrix} \mathbf{A}_d &  \mathbf{B}_d \end{bmatrix}. \\
\end{equation}

Then, to solve for the mass matrix $\mathbf{M}$, we simply take $C_{[d:2d, 2d:3d]}$, multiply by $\frac{1}{dt}$ and invert the result of this step to get $\mathbf{M}$. Then, we can use the part of $\hat{\mathbf{C}}$ that corresponds to $\mathbf{A}_d$ to solve for $\mathbf{C}$ and $\mathbf{K}$. Specifically, we obtain
\begin{equation} \left( \frac{\mathbf{A}_d - \mathbf{I}}{\Delta t} \right) = 
\begin{bmatrix}
\mathbf{0} & \mathbf{I} \\
-\mathbf{M}^{-1}\mathbf{K} & -\mathbf{M}^{-1}\mathbf{C} \\
\end{bmatrix}. \\
\end{equation}

If we further decompose the left hand side of Equation [CITE] using the process described for in Section [CITE], we can obtain specific solutions for $\mathbf{C}$ and $\mathbf{K}$. 

```python
dof = 1

#Decompose C
Ad_approx = C_E[0:2 * dof, 1 * dof - 1:2 * dof]
Bd_approx = C_E[0:2 * dof, 2 * dof : 3 * dof]
print(Ad_approx.shape, Bd_approx.shape)

#Determine M
M = np.linalg.inv(Fs * Bd_approx[dof:2 * dof, 0:dof])
print(M)

#Determine the lower half of (A_d - I) / dT
A_approx = (Ad_approx - np.eye(2 * dof)) * Fs
A_lower = A_approx[dof:2*dof, 0:2*dof]

#Solve the dot product between the lower half and M
LHS = np.dot(M, A_lower)

#Partition into C and K
K = -1 * LHS[0:dof, 0:dof]

C = -1 * LHS[0:dof, dof: 2 * dof]

#Visualise results
print("\nM:", M, "\n\n")
print("K:", K, "\n\n")
print("C:", C, "\n\n")
```

```python
#System properties
def iter_alpha(alpha):
    dof = 1
    m = 1
    c = 10
    
    k = lambda t: 1000 * (1 + alpha * np.sin(2 * np.pi * t))
    t_end = 10
    Fs = 10000
    t = np.arange(0, t_end, 1/Fs)

    #Initial conditions
    x0 = 1e-3
    v0 = -1

    #Forcing function
    def forcing_function(t):

        return 100 * np.cos(2 * np.pi * t)

    #First order DE matrices


    A_mat = lambda t: np.array([[0, 1],
                      [-k(t)/m, -c/m]])

    B_mat = np.array([[0], [1 / m]])

    A_d = lambda dt, t: dt * A_mat(t) + np.eye(2 * dof)
    B_d = lambda dt, t: dt * B_mat
    Ut = lambda t: np.array([[forcing_function(t)]])

    def first_order_DE(t, x):

        return np.dot(A_mat(t), x) + np.dot(B_mat, Ut(t))

    def discrete_DE(x_k, dt, t):

        x_new = np.dot(A_d(dt, t), x_k) + np.dot(B_d(dt, t), Ut(t))

        return x_new

    def Euler(x0, dt, t):

        x0 = x0.reshape(-1, 1)

        Euler_solution = np.zeros((x0.shape[0], len(t)))
        Euler_solution[:, 0] = x0[:, 0]

        for cnt, t_value in enumerate(t[1:], start = 1):

            Euler_solution[:, cnt] = discrete_DE(Euler_solution[:, [cnt - 1]], dt, t_value)[:, 0]

        return Euler_solution

    #Initial conditions
    initial_conditions = np.array([x0, v0]) # [displacement(t=0) = 0, velocity(t=0) = 0]

    #Euler solution
    x_E = Euler(initial_conditions, 1/Fs, t)
    
    ###############################################
    #Estimate the system matrix
    ###############################################
    S_E = np.hstack((x_E.T, Ut(t).reshape(-1, 1))).T

    #Euler
    X_n_E = x_E[:, 1:].T
    S_nm1_E = S_E[:, 0:-1].T
    S_n_E = S_E[:, 1:].T

    ATA_E = np.dot(S_nm1_E.T, S_nm1_E)
    ATB_E = np.dot(S_nm1_E.T, X_n_E)
    C_E = np.transpose(np.linalg.solve(ATA_E, ATB_E))

    #Check the solved for values
    dof = 1

    #Decompose C
    Ad_approx = C_E[0:2 * dof, 1 * dof - 1:2 * dof]
    Bd_approx = C_E[0:2 * dof, 2 * dof : 3 * dof]

    #Determine M
    M = np.linalg.inv(Fs * Bd_approx[dof:2 * dof, 0:dof])

    #Determine the lower half of (A_d - I) / dT
    A_approx = (Ad_approx - np.eye(2 * dof)) * Fs
    A_lower = A_approx[dof:2*dof, 0:2*dof]

    #Solve the dot product between the lower half and M
    LHS = np.dot(M, A_lower)

    #Partition into C and K
    K = -1 * LHS[0:dof, 0:dof]

    C = -1 * LHS[0:dof, dof: 2 * dof]
    
    return M[0, 0], K[0, 0], C[0, 0]


M_alpha = []
C_alpha = []
K_alpha = []
alpha_range = np.linspace(0, 1, 100)

for alpha in alpha_range:
    
    Mi, Ki, Ci = iter_alpha(alpha)
    
    M_alpha.append(Mi)
    C_alpha.append(Ci)
    K_alpha.append(Ki)

fig, ax = plt.subplots(1, 3, figsize = (14, 6))
ax = ax.flatten()

ax[0].plot(alpha_range, M_alpha)
ax[0].set_ylabel("Estimated mass")

ax[1].plot(alpha_range, C_alpha)
ax[1].set_ylabel("Estimated damping coefficient")

ax[2].plot(alpha_range, K_alpha)
ax[2].set_ylabel("Estimated spring stiffness")

for axs in ax:
    axs.set_xlabel(r"$\alpha \in [0, 1]$")
    axs.grid()
    
fig.tight_layout()
plt.savefig("./images/Q4c_fig7.png")
plt.show()
```
